### Setup

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Elapsed Time - Box Plots

```python
# Load the CSV file into a DataFrame
# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv('EE - Experiment 2 (Data) - cleaned_data.csv')

# Parameters
print_outliers = True

# 1. Box plot: Elapsed Time for Kebab vs Camel
plt.figure(figsize=(8, 6))
elapsed_time_df = df[df['Type'].isin(['kebab', 'camel'])]
sns.boxplot(data=elapsed_time_df, x='Type', y='Elapsed Time', palette="Blues", showfliers=print_outliers)
plt.title('Elapsed Time for Kebab Case vs Camel Case')
plt.xlabel('Case Type')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()

# 2. Box plot: Elapsed Time Kebab by Coding Proficiency
plt.figure(figsize=(10, 6))
kebab_df = df[df['Type'] == 'kebab']
sns.boxplot(data=kebab_df, x='Coding Proficiency', y='Elapsed Time', palette="pastel", showfliers=print_outliers)
plt.title('Elapsed Time for Kebab Case by Coding Proficiency')
plt.xlabel('Coding Proficiency')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()

# 3. Box plot: Elapsed Time Camel by Coding Proficiency
plt.figure(figsize=(10, 6))
camel_df = df[df['Type'] == 'camel']
sns.boxplot(data=camel_df, x='Coding Proficiency', y='Elapsed Time', palette="pastel", showfliers=print_outliers)
plt.title('Elapsed Time for Camel Case by Coding Proficiency')
plt.xlabel('Coding Proficiency')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()

# 4. Box plot: Elapsed Time Kebab by English Reading Proficiency
plt.figure(figsize=(10, 6))
sns.boxplot(data=kebab_df, x='English Proficiency', y='Elapsed Time', palette="pastel", showfliers=print_outliers)
plt.title('Elapsed Time for Kebab Case by English Reading Proficiency')
plt.xlabel('English Reading Proficiency')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()

# 5. Box plot: Elapsed Time Camel by English Reading Proficiency
plt.figure(figsize=(10, 6))
sns.boxplot(data=camel_df, x='English Proficiency', y='Elapsed Time', palette="pastel", showfliers=print_outliers)
plt.title('Elapsed Time for Camel Case by English Reading Proficiency')
plt.xlabel('English Reading Proficiency')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()

# 6. Box plot: Elapsed Time Camel by Input Device
plt.figure(figsize=(10, 6))
sns.boxplot(data=camel_df, x='Input Device', y='Elapsed Time', palette="pastel", showfliers=print_outliers)
plt.title('Elapsed Time for Camel Case by Input Device')
plt.xlabel('Input Device')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()

# 8. Box plot: Elapsed Time for English Proficiency by Input Device
plt.figure(figsize=(10, 6))
mouse_device_df = df[df['Input Device'].isin(['mouse'])]
sns.boxplot(data=mouse_device_df, x='Coding Proficiency', y='Elapsed Time', palette="pastel", showfliers=print_outliers)
plt.title('Elapsed Time for Mouse by Coding Proficiency')
plt.xlabel('Coding Proficiency')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()

# 8. Box plot: Elapsed Time for Coding Proficiency by Input Device
plt.figure(figsize=(10, 6))
touchpad_device_df = df[df['Input Device'].isin(['touchpad'])]
sns.boxplot(data=touchpad_device_df, x='Coding Proficiency', y='Elapsed Time', palette="pastel", showfliers=print_outliers)
plt.title('Elapsed Time for Touchpad by Coding Proficiency')
plt.xlabel('Coding Proficiency')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()

# 8. Box plot: Elapsed Time Camel for Case by Input Device 
# Filter for relevant types and input devices
filtered_data = df[(df['Type'].isin(['kebab', 'camel'])) & 
                     (df['Input Device'].isin(['touchpad', 'mouse']))]

# Create a new column for grouping
filtered_data['Group'] = filtered_data['Type'] + ' - ' + filtered_data['Input Device']

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Plot the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=filtered_data, x='Group', y='Elapsed Time', palette='pastel')

# Add title and labels
plt.title('Elapsed Time for Different Cases and Input Devices', fontsize=14)
plt.xlabel('Case and Input Device', fontsize=12)
plt.ylabel('Elapsed Time', fontsize=12)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

### Pie Charts

```python
# Pie charts for English and Coding Proficiency
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load dataset (ensure the correct path if running independently)
data_path = 'EE - Experiment 2 (Data) - cleaned_data.csv'
data = pd.read_csv(data_path)

# Set theme
sns.set_theme(style="whitegrid")

# English Proficiency Pie Chart
english_proficiency = data['English Proficiency'].value_counts(normalize=True) * 100
plt.figure(figsize=(6, 6))
plt.pie(english_proficiency.values, labels=english_proficiency.keys(), autopct='%1.1f%%', 
        startangle=140, colors=sns.color_palette("pastel"))
plt.title('English Proficiency Distribution')
plt.show()

# Coding Proficiency Pie Chart
coding_proficiency = data['Coding Proficiency'].value_counts(normalize=True) * 100
plt.figure(figsize=(6, 6))
plt.pie(coding_proficiency.values, labels=coding_proficiency.keys(), autopct='%1.1f%%', 
        startangle=140, colors=sns.color_palette("pastel"))
plt.title('Coding Proficiency Distribution')
plt.show()
```

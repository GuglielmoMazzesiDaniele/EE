### Setup

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Elapsed Time - Box Plots

```python
# Load the CSV file into a DataFrame
# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv('EE - Experiment 2 (Data) - All Data.csv')

# Parameters
print_outliers = True

# 1. Box plot: Elapsed Time for Kebab vs Camel
plt.figure(figsize=(8, 6))
elapsed_time_df = df[df['Type'].isin(['kebab', 'camel'])]
sns.boxplot(data=elapsed_time_df, x='Type', y='Elapsed Time', palette="Blues", showfliers=print_outliers)
plt.title('Elapsed Time for Kebab Case vs Camel Case')
plt.xlabel('Case Type')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()

# 2. Box plot: Elapsed Time Kebab by Coding Proficiency
plt.figure(figsize=(10, 6))
kebab_df = df[df['Type'] == 'kebab']
sns.boxplot(data=kebab_df, x='Coding Proficiency', y='Elapsed Time', palette="Set2", showfliers=print_outliers)
plt.title('Elapsed Time for Kebab Case by Coding Proficiency')
plt.xlabel('Coding Proficiency')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()

# 3. Box plot: Elapsed Time Camel by Coding Proficiency
plt.figure(figsize=(10, 6))
camel_df = df[df['Type'] == 'camel']
sns.boxplot(data=camel_df, x='Coding Proficiency', y='Elapsed Time', palette="Set3", showfliers=print_outliers)
plt.title('Elapsed Time for Camel Case by Coding Proficiency')
plt.xlabel('Coding Proficiency')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()

# 4. Box plot: Elapsed Time Kebab by English Reading Proficiency
plt.figure(figsize=(10, 6))
sns.boxplot(data=kebab_df, x='English Proficiency', y='Elapsed Time', palette="muted", showfliers=print_outliers)
plt.title('Elapsed Time for Kebab Case by English Reading Proficiency')
plt.xlabel('English Reading Proficiency')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()

# 5. Box plot: Elapsed Time Camel by English Reading Proficiency
plt.figure(figsize=(10, 6))
sns.boxplot(data=camel_df, x='English Proficiency', y='Elapsed Time', palette="pastel", showfliers=print_outliers)
plt.title('Elapsed Time for Camel Case by English Reading Proficiency')
plt.xlabel('English Reading Proficiency')
plt.ylabel('Elapsed Time [ms]')
plt.tight_layout()
plt.show()
```

### Accuracy - Box Plots

```python
# OPTIONAL: Accuracy PLOTS

# 6. Box plot: Accuracy for Kebab vs Camel
plt.figure(figsize=(8, 6))
sns.boxplot(data=elapsed_time_df, x='Type', y='Accuracy', palette="coolwarm")
plt.title('Accuracy for Kebab vs Camel')
plt.xlabel('Algorithm Type')
plt.ylabel('Accuracy (%)')
plt.tight_layout()
plt.show()

# 7. Box plot: Accuracy Kebab by Coding Proficiency
plt.figure(figsize=(10, 6))
sns.boxplot(data=kebab_df, x='Coding Proficiency', y='Accuracy', palette="husl")
plt.title('Accuracy for Kebab by Coding Proficiency')
plt.xlabel('Coding Proficiency')
plt.ylabel('Accuracy (%)')
plt.tight_layout()
plt.show()

# 8. Box plot: Accuracy Camel by Coding Proficiency
plt.figure(figsize=(10, 6))
sns.boxplot(data=camel_df, x='Coding Proficiency', y='Accuracy', palette="YlGnBu")
plt.title('Accuracy for Camel by Coding Proficiency')
plt.xlabel('Coding Proficiency')
plt.ylabel('Accuracy (%)')
plt.tight_layout()
plt.show()

# 9. Box plot: Accuracy Kebab by English Reading Proficiency
plt.figure(figsize=(10, 6))
sns.boxplot(data=kebab_df, x='English Proficiency', y='Accuracy', palette="Oranges")
plt.title('Accuracy for Kebab by English Reading Proficiency')
plt.xlabel('English Reading Proficiency')
plt.ylabel('Accuracy (%)')
plt.tight_layout()
plt.show()

# 10. Box plot: Accuracy Camel by English Reading Proficiency
plt.figure(figsize=(10, 6))
sns.boxplot(data=camel_df, x='English Proficiency', y='Accuracy', palette="Greens")
plt.title('Accuracy for Camel by English Reading Proficiency')
plt.xlabel('English Reading Proficiency')
plt.ylabel('Accuracy (%)')
plt.tight_layout()
plt.show()
```
